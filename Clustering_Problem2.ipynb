```python
# Import necessary libraries
import numpy as np # linear algebra
import pandas as pd # data processing
from matplotlib import pyplot as plt #graphs and plots
from sklearn.cluster import KMeans
%matplotlib inline
```

```python
dfseeds = pd.read_csv("seeds_dataset.txt",sep="\t") #loadtaset
```

```python
dfseeds.shape
```

```python
dfseeds.head()
```

```python
dfseeds.describe()
```

```python
'''To construct the data, seven geometric parameters of wheat kernels were measured:
1. area A,
2. perimeter P,
3. compactness C = 4*pi*A/P^2,
4. length of kernel,
5. width of kernel,
6. asymmetry coefficient
7. length of kernel groove.
All of these parameters were real-valued continuous.
'''
```

```python
dfseeds.columns = ["area", "perimeter", "compactness", "kernel_length","kernel_width","asymmetry_coef","groove_length","class"]
```

```python
dfseeds.head()
```

```python
dfseeds.shape[1]
```

```python
X = dfseeds.values[:, 0:3]
kmeans = KMeans(n_clusters=3, random_state=0).fit(X)
```

```python
X.shape
```

```python
kmeans.labels_
```

```python
kmeans.cluster_centers_
```

```python
'''import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d
fig = plt.figure(figsize = (10, 7))
ax = plt.axes(projection ="3d")'''
%matplotlib inline
```

```python
x= dfseeds["area"]
y = dfseeds["perimeter"]
z= dfseeds["compactness"]
label=dfseeds["class"]
```

```python
# Import libraries
from mpl_toolkits import mplot3d
import numpy as np
import matplotlib.pyplot as plt
colors = ['red','green','blue','purple']
# Create Figure
fig = plt.figure(figsize = (10, 10))
ax = plt.axes(projection ="3d")
# Create Plot
ax.scatter3D(x, y, z,c=label, cmap=matplotlib.colors.ListedColormap(colors))
#Show plot
plt.show()
```

```python
fig2 = plt.figure(figsize = (10, 10))
ax2 = plt.axes(projection ="3d")
# Create Plot
ax2.scatter3D(x,y,z,c=kmeans.labels_, cmap=matplotlib.colors.ListedColormap(colors))
#Show plot
plt.show()
```
